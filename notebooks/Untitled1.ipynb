# Compare monomer and multimer clustering 

```python
# imports
import glob 
import pandas as pd 
from collections import Counter
import re
```

```python
# get the multimer cluster reports 
multimer_reports = glob.glob('/home/grig0076/scratch/phlegm/PHROGs/phrog_representative_structures/phrog_predicted_oligomers/foldseek_multimercluster/*.tsv')
multimer_cluster_df = pd.DataFrame()

for m in multimer_reports: 
    tm_threshold = re.split('_',m)[6]
    coverage = re.split('_',m)[8] 
    report = pd.read_csv(m, sep='\t', header = None)
    cluster_sizes = pd.DataFrame.from_dict(Counter(report[0]), orient='index') 
    num_clusters = len(cluster_sizes)
    num_singletons = len(cluster_sizes[cluster_sizes[0] == 1] )
    df =   pd.DataFrame.from_dict({'path': m, 'tm_threshold': tm_threshold, 'coverage': coverage, 'num_clusters': num_clusters, 'num_singletons': num_singletons}, orient='index').T
    multimer_cluster_df = pd.concat([multimer_cluster_df, df])
```
