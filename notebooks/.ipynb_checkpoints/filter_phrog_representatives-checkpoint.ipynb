```python
# imports 
import pandas as pd 
from Bio import SeqIO
import seaborn as sns 
import matplotlib.pyplot as plt
import numpy as np
import glob
from Bio.PDB import PDBParser
import re
from scipy.stats import gaussian_kde
from collections import Counter
```

## Filter by sequence length

```python
# import the sequences so can see how long they are 
fasta_dict = {record.id: str(record.seq) for record in SeqIO.parse("../files/nonsingleton_representative_sequences.fasta", "fasta")}

# get the length of each phrog 
seq_lengths = [len(f) for f in list(fasta_dict.values())]

# plot the distribution of sequence lengths 
sns.set(font_scale=1.4,style='white')
sns.displot(seq_lengths)
plt.xlabel('Sequence Length (Amino Acids)')
plt.axvline(x=350, color='black', linestyle='--')
plt.savefig('../figures/PHROG_representative_lengths.png',  dpi=300, bbox_inches='tight')
```

```python
# Apply filtering to take sequences with a length less than 350
phrogs_length_filter = np.array(list(fasta_dict.keys()))[np.array(seq_lengths) <= 350] 
phrogs_length_remove = np.array(list(fasta_dict.keys()))[np.array(seq_lengths) > 350] 
```

```python
# read in the anotations of those not included 
phrog_annot = pd.read_csv('../files/phrog_annot_v4.tsv', sep = '\t')
phrog_category = dict(zip(['phrog_' + str(p)   for p in phrog_annot['phrog']], phrog_annot['category']))
phrog_annot = dict(zip(['phrog_' + str(p) for p in phrog_annot['phrog']], phrog_annot['annot']))
```

```python
# Count annotations
long_phrog_annotations = Counter([phrog_annot.get(p) for p in phrogs_length_remove])

# Filter annotations with count > 10 and exclude NaNs
filtered_annotations = {
    label: count for label, count in long_phrog_annotations.items()
    if count > 15 and label == label  # Excludes NaN
}

# Sort annotations by count (descending)
sorted_items = sorted(filtered_annotations.items(), key=lambda x: x[1], reverse=True)

# Extract labels and counts
labels = [str(label) for label, _ in sorted_items]
counts = [count for _, count in sorted_items]

# Create a bar plot
plt.figure(figsize=(10, 6))
plt.barh(labels, counts, color='skyblue')

# Add labels and title
plt.xlabel('Count')
plt.ylabel('Annotation')

# Adjust layout to fit labels
plt.tight_layout()
plt.savefig('../figures/excluded_annotations.png',  dpi=300, bbox_inches='tight')
plt.show()
```

## Filter by the plddt of the folded monomer

```python
plddt_dict = dict()

def get_mean_plddt(pdb_file):
    parser = PDBParser(QUIET=True)
    structure = parser.get_structure("model", pdb_file)

    plddt_scores = []

    for model in structure:
        for chain in model:
            for residue in chain:
                for atom in residue:
                    if atom.get_name() == "CA":  # Use only CA atoms for per-residue scores
                        plddt_scores.append(atom.get_bfactor())
                        break  # Only need one B-factor per residue

    if not plddt_scores:
        return None  # Or raise an error if preferred

    return sum(plddt_scores) / len(plddt_scores)

for m in monomers: 
    mean_plddt = get_mean_plddt(m)
    plddt_dict['_'.join(re.split('_', re.split('/', m)[-1])[:2])] = mean_plddt
```

```python
plddt_scores = [plddt_dict.get(p) for p in phrogs_length_filter if p in plddt_dict.keys()]

# Set up figure
plt.figure(figsize=(11, 5))
sns.set(font_scale=1.5, style='white')

# Plot the KDE once for the full distribution
sns.kdeplot(plddt_scores, fill=False, color="black", linewidth=2)

# Generate KDE values for shading
x_vals = np.linspace(0, 100, 500)
kde = gaussian_kde(plddt_scores)
y_vals = kde(x_vals)

# Define intervals in descending order for reversed legend
intervals = [
    (90, 100, "#1f4dc2", "Very high (pLDDT > 90)"),
    (70, 90, "#67c2ed", "High (90 > pLDDT > 70)"),
    (50, 70, "#ffd700", "Low (70 > pLDDT > 50)"),
    (0, 50, "#f07e6e", "Very low (pLDDT ≤ 50)"),
]

# Plot shaded regions
for xmin, xmax, color, label in intervals:
    mask = (x_vals >= xmin) & (x_vals < xmax)
    plt.fill_between(x_vals[mask], y_vals[mask], color=color, alpha=0.6, label=label)

# Final formatting
plt.xlabel("pLDDT Score")
plt.ylabel("Density")
plt.xlim(0, 100)

# Move legend outside
plt.legend(
    bbox_to_anchor=(1.05, 1),
    loc="upper left",
    borderaxespad=0,
)
plt.tight_layout()
plt.savefig('../figures/PHROG_representative_plddt.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# save the phrogs that will be used for analysis in the paper 
phrogs_pass =  list(phrogs_length_filter[np.array(plddt_scores) >= 70 ])

# save this info to a file 
with open('../files/phrogs_pass_plddt_length_threshold.txt', 'w') as file:
    for item in phrogs_pass:
        file.write(f"{item}\n")
```

## plot a comparison of the PLDDT and the Neff and the number of sequences in the MSA 

```python
# read in the neff 
neff_dict = {} 
msadepth_dict = {} 

neff_files = glob.glob('/home/grig0076/scratch/phlegm/PHROGs/phrog_representative_structures/monomers/neff/*')
for n in neff_files: 
    neff_df = pd.read_csv(n, sep =':', header=None, index_col=0)
    
    neff_dict[re.split('\/', n)[-1][:-8]] = neff_df.loc['NEFF'][1] 
    msadepth_dict[re.split('\/', n)[-1][:-8]] = neff_df.loc['MSA depth'][1] 
```

```python
metrics_df = pd.concat([pd.DataFrame.from_dict(neff_dict, orient='index'),  pd.DataFrame.from_dict(msadepth_dict, orient='index'),  pd.DataFrame.from_dict(plddt_dict, orient='index')], axis=1)
metrics_df = metrics_df[metrics_df.index.isin(phrogs_pass)] # filter to only include some of them 
metrics_df.columns = ['NEFF', 'MSA depth', 'pLDDT'] 
```

```python


# Drop NaNs
df = metrics_df.dropna(subset=['pLDDT', 'NEFF', 'MSA depth'])

# Log-transform
df['log_neff'] = np.log10(df['NEFF'])
df['log_msadepth'] = np.log10(df['MSA depth'])

sns.set(style='white', font_scale=1.6)

# --- log(NEFF) vs pLDDT ---
plt.figure(figsize=(6, 6))
sns.kdeplot(
    data=df,
    x='log_neff',
    y='pLDDT',
    fill=True,
    cmap='Blues',
    thresh=0.01,
    levels=100,
)
sns.regplot(
    data=df,
    x='log_neff',
    y='pLDDT',
    scatter=False,
    color='black',
    line_kws={'linewidth': 2}
)
plt.xlabel("log10(NEFF)")
plt.ylabel("Mean pLDDT")
#plt.ylim(0, 100)
plt.tight_layout()
plt.savefig('../figures/neff_vs_plddt.png', dpi=300, bbox_inches='tight')
plt.show()

# --- log(MSA depth) vs pLDDT ---
plt.figure(figsize=(6, 6))
sns.kdeplot(
    data=df,
    x='log_msadepth',
    y='pLDDT',
    fill=True,
    cmap='Blues',
    thresh=0.01,
    levels=100,
)
sns.regplot(
    data=df,
    x='log_msadepth',
    y='pLDDT',
    scatter=False,
    color='black',
    line_kws={'linewidth': 2}
)
plt.xlabel("log10(MSA depth)")
plt.ylabel("Mean pLDDT")
#plt.ylim(0, 100)
plt.tight_layout()
plt.savefig('../figures/msadepth_vs_plddt.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# compute stats on these 

import numpy as np
from scipy.stats import pearsonr, spearmanr

# Drop rows with NaNs in relevant columns
df = metrics_df.dropna(subset=['pLDDT', 'NEFF', 'MSA depth'])

# Optionally log-transform neff and msadepth for Pearson correlation
df['log_neff'] = np.log10(df['NEFF'])
df['log_msadepth'] = np.log10(df['MSA depth'])

# Pearson correlation (linear relationship)
pearson_neff = pearsonr(df['log_neff'], df['pLDDT'])
pearson_msadepth = pearsonr(df['log_msadepth'], df['pLDDT'])

# Spearman correlation (rank-based)
spearman_neff = spearmanr(df['NEFF'], df['pLDDT'])
spearman_msadepth = spearmanr(df['MSA depth'], df['pLDDT'])

print(f"Pearson correlation (log neff vs pLDDT): r={pearson_neff[0]:.3f}, p={pearson_neff[1]:.2e}")
print(f"Spearman correlation (neff vs pLDDT): r={spearman_neff[0]:.3f}, p={spearman_neff[1]:.2e}")
print(f"Pearson correlation (log msadepth vs pLDDT): r={pearson_msadepth[0]:.3f}, p={pearson_msadepth[1]:.2e}")
print(f"Spearman correlation (msadepth vs pLDDT): r={spearman_msadepth[0]:.3f}, p={spearman_msadepth[1]:.2e}")
```
